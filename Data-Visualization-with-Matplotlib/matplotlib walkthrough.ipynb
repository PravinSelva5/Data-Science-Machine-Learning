```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# FUNCTIONAL
plt.plot(x,y)
# plt.show() would be needed if you weren't using juptyer notebook
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

# MatplotLib Part 2

```python
fig,axes = plt.subplots(nrows=1, ncols=2)  # tuple unpacking, axes is an array of matplotlib.axes

'''
for current_ax in axes:
    current_ax.plot(x,y)
'''

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

#axes.plot(x,y)
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2)) # dpi = dots per inch or pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Exporting / Saving Information

```python
fig.savefig('my_picture.png', dpi=200) # the higher the DPI value, the more detail seen in the chart
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc='best')
```

# Matplotlib - Part 3

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', linewidth=1.5, linestyle='--',marker='o', markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')  # color='#FF8C00', takes in RGB Hex Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_xlim([0,1])
ax.set_ylim([0,2])

ax.plot(x, y)
```
